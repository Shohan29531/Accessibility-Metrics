```python

data = pd.read_csv('pairwise_data_our_exp.csv')

print(data.head())


# data['src_dest'] = data['source'].astype(str)

# data['src_dest'] = data['src_dest'].apply( lambda x: x + '_' + data['destination'].astype(str))

dummy = []

for i in range(len(data['source'])):
  dummy.append(str(data['source'][i]) + "_" + str(data['destination'][i]))

data['src_dest'] = dummy

display(data.head())

```
